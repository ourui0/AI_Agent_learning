```python
# 定义目标函数（模拟损失函数）
def f(x):
    return x**2

# 定义导函数
def grad(x):
    return 2*x
```

```python
# 用列表保存点的变化轨迹
x_list = []
y_list = []
```

```python
# 初始化参数和超参数
x = 1
alpha = 0.1
```

```python
for i in range(100):
    # 记录当前点坐标
    x_list.append(x)
    y_list.append(f(x))
    print(f"x :{x},\tf(x) = {f(x)},\t")

    # 计算梯度
    g = grad(x)

    # 更新参数
    x = x - alpha * g
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-1, 1, 0.01)
plt.plot(x, f(x))
plt.plot(x_list, y_list,"r")
plt.scatter(x_list, y_list, c="r")
plt.show()
```
