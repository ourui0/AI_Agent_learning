```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]) # 自变量，每周学习时长
y = np.array([[55], [65], [70], [75], [85], [50], [60], [72], [80], [58]]) # 因变量，数学考试成绩
```

```python
x = np.array(X)
y = np.array(y).reshape(-1, 1)
n = x.shape[0]
# 增加一列
x = np.hstack([np.ones((n,1)),x])
print(x)
print(x.shape)
print(y.shape)
```

```python
# 定义目标函数（损失函数）
def J(beta):
    return np.sum((x @ beta - y)**2)/n
```

```python
# 定义梯度计算方法
def gradient(beta):
    return x.T @ (x @ beta - y) / n * 2
```

```python
# 初始化参数和超参数
beta = np.array([1,1]).reshape(-1,1)
alpha = 0.01
iter = 10000
```

```python
# 重复迭代
beta0 = []
beta1 = []
# for i in range(iter):
while(np.abs(grad := gradient(beta)) > 1e-6).any() and (iter := iter - 1) >= 0:
    beta0.append(beta[0,0])
    beta1.append(beta[1,0])
    # 计算梯度
    # grad = gradient(beta)
    # 更新参数
    beta = beta - alpha * grad
    if iter % 10 == 0:
        print(f"beta:{beta.reshape(-1)}\tJ:{J(beta)}\tgrad:{grad}")
print(iter)
```

```python
# 画图
plt.plot(beta0, beta1)
plt.show()
```
