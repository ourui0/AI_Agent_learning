```python
import numpy as np
```

```python
# 生成特征
x1 = np.random.randn(100)
print(np.var(x1))
print(x1)
```

```python
x2 = np.random.normal(5,0.1,100)
print(np.var(x2))
print(x2)
```

```python
#将特征组合为数据矩阵
x = np.vstack((x1,x2)).T
print(x.shape)
```

```python
#特征选择：低方差过滤法
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.01)
```

```python
x_filtered = vt.fit_transform(x)
print(x_filtered)
print(x_filtered.shape)
```
