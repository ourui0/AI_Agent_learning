```python
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
label = ["猫","狗"] # 标签
y_true = ["猫","猫","猫","猫","猫","猫","狗","狗","狗","狗"] # 真实组
y_pred = ["猫","猫","猫","狗","猫","猫","猫","猫","狗","狗"] # 预测组
```

```python
matrix = confusion_matrix(y_true,y_pred,labels = label)
print(matrix)
```

```python
print(pd.DataFrame(matrix,columns=label,index = label))
```

```python
import seaborn as sns
sns.heatmap(matrix,annot = True,fmt = "d",cmap= "Greens")
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true,y_pred)
print(accuracy)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_true,y_pred,pos_label="猫")
print(precision)
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_true,y_pred,pos_label="猫")
print(recall)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_true,y_pred,pos_label="猫")
print(f1)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_true,y_pred,labels = label,target_names= None)
print(report)
```
