```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# 定义特征
x = [[2,1],[3,1],[1,4],[2,6]]
```

```python
# 定义归一化缩放器
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
# 定义特征
x = [[2,1],[3,1],[1,4],[2,6]]
```

```python
# 定义标准化缩放器
scaler = StandardScaler()
```

```python
# 标准化处理
x = scaler.fit_transform(x)
print(x)
```
