```python
from IPython.core.pylabtools import figsize


# 定义目标函数（模拟损失函数）
def f(x):
    return (x**2 - 2)**2

# 定义导函数
def grad(x):
    return 4 * x**3 - 8 * x
```

```python
# 用列表保存点的变化轨迹
x_list = []
y_list = []
```

```python
# 初始化参数和超参数
x = 1
alpha = 0.1
```

```python
import numpy as np
# for i in range(100):
while np.abs(g := grad(x)) > 1e-6:
    # 记录当前点坐标
    x_list.append(x)
    y_list.append(f(x))
    print(f"x :{x},\tf(x) = {f(x)},\t")

    # 计算梯度
    # g = grad(x)

    # 更新参数
    x = x - alpha * g
```

```python
import matplotlib.pyplot as plt
x = np.arange(0.9, 1.6, 0.01)
plt.plot(x, f(x))
plt.plot(x_list, y_list,"r")
plt.scatter(x_list, y_list, c="r")
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,4))
x = np.arange(-2, 2, 0.01)
ax[0].plot(x, f(x))
ax[0].plot(x_list, y_list,"r")
ax[0].scatter(x_list, y_list, c="r")

x2 = np.arange(1.39,1.425,0.001)
x_list2 = x_list[1:]
y_list2 = y_list[1:]
ax[1].plot(x2, f(x2))
ax[1].plot(x_list2, y_list2,"r")
ax[1].scatter(x_list2, y_list2, c="r")
plt.show()
```
