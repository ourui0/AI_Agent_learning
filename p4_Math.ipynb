```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#画出y=sin(x)图像
x = np.arange(0,6,0.1)
y = np.sin(x)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("y = sin(x)")
plt.show()
```

```python
y1 = np.cos(x)
plt.plot(x,y1,'y--',label = "y = cos(x)")
plt.plot(x,y,label = "y = sin(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title("y = cos(x)")
plt.legend()
plt.show()
```

```python
x = np.array([2,3,4])
y = np.array([5,6,7])
print(x * y)#这是对位相乘，并不是向量点积
print(x.dot(y))
print(np.dot(x,y))
print(x @ y)
```

```python
l0_norm = np.linalg.norm(x,ord = 0)
print(l0_norm)
l1_norm = np.linalg.norm(x,ord = 1)
print(l1_norm)
lp_norm = np.linalg.norm(x,ord = np.inf)
print(l0_norm)
```

```python
A = np.array([[1,2,3],[4,5,6]])
print(A + x)#x会被广播
print(A * x)#依旧是对位相乘
print(A.dot(x))
print(np.dot(A,x))
print(A @ x)
```

```python
D = np.array([[1,2,3],[4,5,6],[7,8,10]])
D_inv = np.linalg.inv(D)
print(D_inv)
print(D @ D_inv)#部分不直接为0，浮点数计算精度问题
result = D @ D_inv
print(np.around(result).astype(int))
```
