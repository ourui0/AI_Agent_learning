```python
import torch
import torch.nn as nn
```

```python
# 定义输入数据和初始隐状态
input = torch.randn(1, 3, 8)
hx = torch.randn(2, 3, 16)
```

```python
# 定义rnn层
rnn = nn.RNN(input_size = 8, hidden_size = 16, num_layers = 2)
```

```python
# 前向传播
output, hn = rnn(input, hx)
```

```python
print(output)
print(hn)
```
