```python
import torch
from torch.nn import MSELoss,L1Loss,SmoothL1Loss
```

```python
# 定义数据
input = torch.randn(3,5)
target = torch.randn(3,5)
print(input)
print(target)
```

```python
# MAE
l1_loss = L1Loss()
l1 = l1_loss(input, target)
print(l1)
```

```python
# MSE
mse_loss = MSELoss()
mse = mse_loss(input, target)
print(mse)
```

```python
# SmoothL1
SmoothL1_loss = SmoothL1Loss()
SmoothL1 = SmoothL1_loss(input, target)
print(SmoothL1)
```
