```python
import torch
```

```python
# 定义数据
x = torch.tensor(10.0)
y = torch.tensor(3.0)
print(x.requires_grad)
print(y.requires_grad)
```

```python
w = torch.rand(1,1,requires_grad=True)
b = torch.rand(1,1,requires_grad=True)
```

```python
# 得到前向传播的输出
z = x * w + b
print(z)
print(z.requires_grad)
```

```python
# 查看是否为叶子节点
print(x.is_leaf)
print(y.is_leaf)
print(w.is_leaf)
print(z.is_leaf)
print(b.is_leaf)
```

```python
# 设置损失函数
loss = torch.nn.MSELoss()

# 计算损失
loss_value = loss(z,y.view_as(z))
print(loss_value)
print(loss_value.is_leaf)
```

```python
loss_value.backward()
```

```python
print(w.grad)
print(b.grad)
```
